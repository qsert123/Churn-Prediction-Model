```python
#import the lib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
data = pd.DataFrame(data)
data
```

```python
data.info()
```

```python
data = data.drop(columns=["customerID"], axis=1)
```

```python
data
```

```python
##change the  gender column to 0 and 1
data["gender"] = data["gender"].map({"Female": 0, "Male": 1})
data
```

```python
data["Partner"] = data["Partner"].map({"No": 0, "Yes": 1})
data
```

```python
d = data["Dependents"].unique()
d
```

```python
data["Dependents"] = data["Dependents"].map({"No": 0, "Yes": 1})
data
```

```python
data["PhoneService"] = data["PhoneService"].map({"No": 0, "Yes": 1})
data
```

```python
d = data["InternetService"].unique()
d
```

```python
#DSL = 0, Fiber optic = 1, No = 2
data["InternetService"] = data["InternetService"].map({"DSL": 0, "Fiber optic": 1, "No": 2})
data
```

```python
data.drop(columns=["TotalCharges","PaymentMethod"], inplace=True, errors='ignore')
data
```

```python
data.drop(columns=["OnlineSecurity", "OnlineBackup", "DeviceProtection", "TechSupport", "StreamingTV", "StreamingMovies","PaperlessBilling"], inplace=True, errors='ignore')
```

```python
data["MultipleLines"] = data["MultipleLines"].map({"No phone service": 0, "No": 1, "Yes": 2})
data
```

```python
data.info()
```

```python
data["Contract"] = data["Contract"].map({"Month-to-month": 0, "One year": 1, "Two year": 2})
data["Churn"] = data["Churn"].map({"No": 0, "Yes": 1})
```

```python
data
```

```python
data.to_csv("cleaned_data.csv", index=False)
```
